## Ejercicio 1
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']

````python
# Listas dadas
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]

# Crear una nueva lista con las combinaciones de las dos listas
combinaciones = [a + " " + b for a in ejer_3_1 for b in ejer_3_2]

# Imprimir la lista resultante
print(combinaciones)
````

## Ejercicio 3

Crea una función que permita eliminar los elementos de una lista o tupla, incluso los repetidos. Para ello la función debe admitir un número arbitrario de elementos a eliminar, la lista o tupla que debe eliminarse, el número de elementos repetidos que debe eliminarse, este parámetro por defecto debe estar a None, que quiere decir todos los elementos repetidos.

Si el elemento no está en la lista o tupla debe imprimirlo por pantalla y no dar error.


````python
# Función para eliminar elementos de una lista o tupla
def eliminar_elementos(coleccion, elementos_a_eliminar, numero_a_eliminar=None):
    # Convertir la tupla en lista si es necesario, ya que las tuplas son inmutables
    if isinstance(coleccion, tuple):
        coleccion = list(coleccion)
    
    # Si el elemento no está en la lista o tupla, imprimir un mensaje y salir
    for elemento in elementos_a_eliminar:
        if elemento not in coleccion:
            print(f"El elemento {elemento} no está en la colección.")
            continue  # Saltar al siguiente elemento si no se encuentra

        # Si se especifica un número de elementos a eliminar
        if numero_a_eliminar is not None:
            for _ in range(numero_a_eliminar):
                if elemento in coleccion:
                    coleccion.remove(elemento)
                else:
                    break
        # Si no se especifica, eliminar todos los elementos repetidos
        else:
            while elemento in coleccion:
                coleccion.remove(elemento)

    # Devolver la colección modificada
    return coleccion

# Ejemplos de uso

# Ejemplo con una lista
lista = [1, 2, 3, 3, 4, 5, 3, 6, 3]
print("Lista original:", lista)
nueva_lista = eliminar_elementos(lista, [3], numero_a_eliminar=2)
print("Lista después de eliminar 2 ocurrencias del número 3:", nueva_lista)

# Ejemplo con una tupla
tupla = (1, 2, 3, 3, 4, 5, 3, 6, 3)
print("\nTupla original:", tupla)
nueva_tupla = eliminar_elementos(tupla, [3])
print("Tupla después de eliminar todas las ocurrencias del número 3:", nueva_tupla)

# Ejemplo con un elemento que no está en la colección
lista = [1, 2, 3, 4, 5]
print("\nLista original:", lista)
nueva_lista = eliminar_elementos(lista, [7])  # El 7 no está en la lista
print("Lista después de intentar eliminar el número 7:", nueva_lista)
````

## Ejercicio 4

Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

````python
# Función para añadir o eliminar elementos en una lista
def modificar_lista(lista, comando, elemento=None):
    # Si el comando es "add", añadir el elemento a la lista
    if comando == "add":
        if elemento is not None:  # Verificar que el elemento no sea None
            lista.append(elemento)
            print(f"Elemento '{elemento}' añadido a la lista.")
        else:
            print("No se puede añadir un elemento None.")
    
    # Si el comando es "remove", eliminar el elemento de la lista
    elif comando == "remove":
        if elemento in lista:
            lista.remove(elemento)
            print(f"Elemento '{elemento}' eliminado de la lista.")
        else:
            print(f"El elemento '{elemento}' no está en la lista.")
    
    # Comando no válido
    else:
        print("Comando no reconocido. Usa 'add' o 'remove'.")
    
    return lista

# Ejemplo de uso
mi_lista = [1, 2, 3, 4, 5]

# Añadir un elemento
modificar_lista(mi_lista, "add", 6)
print("Lista después de añadir:", mi_lista)

# Eliminar un elemento
modificar_lista(mi_lista, "remove", 3)
print("Lista después de eliminar:", mi_lista)

# Intentar eliminar un elemento que no está en la lista
modificar_lista(mi_lista, "remove", 10)

# Intentar añadir un elemento None
modificar_lista(mi_lista, "add")
````

## Ejercicio 5

Dividir entre 100 los elementos positivos de una lista

````python
L = [10, 5, -1, 20, 30, -100, 0]

# Método tradicional
# Lista dada
L = [10, 5, -1, 20, 30, -100, 0]

# Método tradicional
resultado_tradicional = []
for elemento in L:
    if elemento > 0:
        resultado_tradicional.append(elemento / 100)

print("Resultado con método tradicional:", resultado_tradicional)
````

````python
# Con comprensión de listas
# Con comprensión de listas
resultado_comprension = [elemento / 100 for elemento in L if elemento > 0]

print("Resultado con comprensión de listas:", resultado_comprension)
````

## Ejercicios 6

Define una función que tenga como entrada una lista de temperaturas y un parametro de temperatura mínima. La fución debe devolver una lista con los elementos de la lista de temperaturas original que sean menores o iguales a la temperatura mínima pasada como parámetro.

Prueba tu función con la siguiente lista:

         
> ```temperaturas = [25.2, 24.9, 25.2, 26.7, 28.6, 29.5, 29.7]```

````python
# Definir la función que filtra las temperaturas
def filtrar_temperaturas(temperaturas, temp_minima):
    # Usamos una comprensión de listas para devolver las temperaturas <= temp_minima
    return [temp for temp in temperaturas if temp <= temp_minima]

# Lista de temperaturas dada
temperaturas = [25.2, 24.9, 25.2, 26.7, 28.6, 29.5, 29.7]

# Prueba de la función con una temperatura mínima de 26.0
resultado = filtrar_temperaturas(temperaturas, 26.0)

print("Temperaturas menores o iguales a 26.0:", resultado)
````

## Ejercicio 7

Repite el ejercicio 6 pero utilizando comprensión de listas para producir la lista resultado. Si ya lo hiciste así en el ejercicio 6, hazlo aquí empleando un bucle ```for``` externo.

````python
# Definir la función que filtra las temperaturas usando un bucle for
def filtrar_temperaturas_bucle(temperaturas, temp_minima):
    resultado = []
    for temp in temperaturas:
        if temp <= temp_minima:
            resultado.append(temp)
    return resultado

# Lista de temperaturas dada
temperaturas = [25.2, 24.9, 25.2, 26.7, 28.6, 29.5, 29.7]

# Prueba de la función con una temperatura mínima de 26.0
resultado = filtrar_temperaturas_bucle(temperaturas, 26.0)

print("Temperaturas menores o iguales a 26.0:", resultado)
````
